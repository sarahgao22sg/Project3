
## Import CSV's



```python


file_path = os.path.join("Resources", "Exotic -Greenhouse Frog.csv")
frogs_df = pd.read_csv(file_path)

frogs_df.head()
```

```python

file_path = os.path.join("Resources", "Exotic - Zebra Mussels.csv")
zebras_df = pd.read_csv(file_path)

zebras_df[["Month", "Day", "Year"]]
```


### Structure Data


```python
wanted_columns = ["Specimen Number", "Species ID", "Group", "Family", "Scientific Name", "Common Name", "Country", "State", "County", "Locality", "Latitude", "Longitude", "Drainage Name", "Year", "Month", "Day"]
zebras_df = zebras_df[wanted_columns]
zebras_df.head()
```

```python
zebras_df = zebras_df.fillna(1)
zebras_df.head()
```

```python
zebras_df['Recorded Date'] = zebras_df.Month.astype(str) + zebras_df.Day.astype(str) + zebras_df.Year.astype(str)
zebras_df.head()
```

```python
zebras_df.drop(['Year', 'Month', 'Day'], axis='columns', inplace=True)
```
