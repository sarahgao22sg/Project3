```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from geopy.geocoders import Nominatim
```

```python
file_path = os.path.join("Resources", "zebra_frogs.csv")
zebras_frogs = pd.read_csv(file_path)

zebras_frogs.head()
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(zebras_frogs["Latitude"], zebras_frogs["Longitude"])
```

```python
zf_map = folium.Map(location=[45.5236, -95.6750], zoom_start=5)
zebras_frogs.apply(lambda row:folium.Marker(location=[row["Latitude"], row["Longitude"]] ).add_to(zf_map), axis=1)
zf_map
```

```python
zebras_frogs["Latitude"] = np.radians(zebras_frogs["Latitude"])
zebras_frogs["Longitude"] = np.radians(zebras_frogs["Longitude"])
```

```python
from sklearn.cluster import KMeans ,AgglomerativeClustering

# # creates 5 clusters using hierarchical clustering.
# agc = AgglomerativeClustering(n_clusters =5, affinity='euclidean', linkage='ward')
# zebras_frogs['pickup cluster'] = agc.fit_predict(zebras_frogs[['Latitude','Longitude']])

# creates 5 clusters using k-means clustering algorithm.
kmeans = KMeans(5)
clusters = kmeans.fit_predict(zebras_frogs[['Latitude','Longitude']])
zebras_frogs['pickup cluster'] = kmeans.predict(zebras_frogs[['Latitude','Longitude']])
```
